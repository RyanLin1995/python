```python
from sklearnex import patch_sklearn
import matplotlib.pyplot as plt
from sklearn import datasets

from sklearn.cluster import KMeans  # 聚类算法api
from sklearn.metrics import calinski_harabasz_score

patch_sklearn()
```

```python
# 获取数据
# 创建数据集
# X为样本特征，Y为样本簇类别， 共1000个样本，每个样本4个特征，共4个簇，
# 簇中心在[-1,-1], [0,0],[1,1], [2,2]， 簇方差分别为[0.4, 0.2, 0.2, 0.2]
x, y = datasets.make_blobs(n_samples=1000, n_features=2, cluster_std=[0.4, 0.1, 0.1, 0.1],
                           centers=[[-1, -1], [0, 0], [1, 1], [2, 2]], random_state=1)
plt.scatter(x[:, 0], x[:, 1], marker='o')
```

```python
# 模型训练
estimator = KMeans(n_clusters=2, random_state=2)  # n_clusters 开始聚类的中心数量
y_pre = estimator.fit_predict(x)
# 结果可视化
plt.scatter(x[:, 0], x[:, 1], c=y_pre)
plt.show()
```

```python
# 模型评估
# 用Calinski-Harabasz Index评估的聚类分数
print(calinski_harabasz_score(x, y_pre))
```
