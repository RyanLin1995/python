```python
import pandas as pd
```

```python
data = pd.read_csv('stock_day.csv')
```

```python
data
```

```python
data = data.drop(["ma5","ma10","ma20","v_ma5","v_ma10","v_ma20"], axis=1)
```

# 1 索引操作

```python
data['open']['2018-02-27']  # 先列后行
```

```python
data.loc['2018-02-27':'2018-02-14', 'open']  # 通过 loc 指定行列索引的名称获取数据
```

```python
data.iloc[:3, :5]  # 通过 iloc 指定行列索引的下标获取数据
```

# 2 赋值操作

```python
data
```

# 2 赋值操作

```python
data['close'] = 100  # 直接整列修改原来的值
```

```python
data
```

```python
data.close = 1  # 直接整列修改原来的值，效果与 data['close'] 一致
```

```python
data
```

# 3 排序

### 1 Dataframe 排序

```python
data.sort_values(by='open', ascending=True).head()  # by：指定排序参考的键，可以是字符串或列表，列表时为对多个键排序;
# ascending指定按照大小排序，默认为升序(Ture)
```

```python
data.sort_values(by=['open', 'high'])  # 对多个键进行排序
```

```python
data.sort_index()  # 对索引进行排序（默认行为，默认为降序）
```

### 2 Series 排序

```python
data['price_change'].sort_values(ascending=True)  # 对 Series 的值进行排序
```

```python
data['p_change'].sort_index().head()  # 对 Series 的索引进行排序
```
