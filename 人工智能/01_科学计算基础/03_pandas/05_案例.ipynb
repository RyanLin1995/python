#### 问题2：对于这一组电影数据，如果我们想rating，runtime的分布情况，应该如何呈现数据？

##### Rating 分布

```python
plt.figure(figsize=(20,8),dpi=80)
plt.hist(movie["Rating"].values,bins=20)  # 绘制直方图

# 求出最大最小值
max_ = movie["Rating"].max()
min_ = movie["Rating"].min()

# 生成刻度列表
t1 = np.linspace(min_,max_,num=21)

# [ 1.9    2.255  2.61   2.965  3.32   3.675  4.03   4.385  4.74   5.095  5.45   5.805  6.16   6.515  6.87   7.225  7.58   7.935  8.29   8.645  9.   ]

# 修改刻度
plt.xticks(t1)

# 添加网格
plt.grid()

plt.show()
```

##### Runtime(Minutes) 分布

```python
plt.figure(figsize=(20,8),dpi=80)
plt.hist(movie["Runtime (Minutes)"].values,bins=20)  # 绘制直方图

# 求出最大最小值
max_ = movie["Runtime (Minutes)"].max()
min_ = movie["Runtime (Minutes)"].min()

# 生成刻度列表
t1 = np.linspace(min_,max_,num=21)

# [ 1.9    2.255  2.61   2.965  3.32   3.675  4.03   4.385  4.74   5.095  5.45   5.805  6.16   6.515  6.87   7.225  7.58   7.935  8.29   8.645  9.   ]

# 修改刻度
plt.xticks(t1)

# 添加网格
plt.grid()

plt.show()
```

#### 问题3：对于这一组电影数据，如果我们希望统计电影分类(genre)的情况，应该如何处理数据？

* 思路分析:
    * 1. 创建一个全为0的dataframe，列索引置为电影的分类，temp_df
    * 2. 遍历每一部电影，temp_df中把分类出现的列的值置为1
    * 3. 求和

```python
temp_list = [i.split(',') for i in movie['Genre']]
```

```python
temp_list
```

```python
array_list = np.unique([j for i in temp_list for j in i])
```

```python
array_list
```

```python
temp_df = pd.DataFrame(np.zeros([movie.shape[0],array_list.shape[0]]),columns=array_list)  # 用 np.zeros 构建一个列名为 array_list 的，值为 0 的空 Dataframe
```

```python
temp_df
```

```python
for i in range(len(temp_df)):
    temp_df.loc[i, temp_list[i]] = 1  # 利用 loc 通过行列修改数值
```

```python
temp_df
```

```python
temp_df.sum().sort_values().plot(kind='bar',figsize=(20,8),fontsize=20,colormap="cool")
```
