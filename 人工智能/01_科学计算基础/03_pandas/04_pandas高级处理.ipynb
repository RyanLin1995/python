# 3 数据离散化

```python
data = pd.read_csv('stock_day.csv')
```

```python
p_change = data['p_change']
```

```python
p_change
```

```python
q_cut = pd.qcut(p_change, q=10)  # 对数据进行分组。q：指定分位数，可以设置百分比或分位数的数，然后根据分位数进行分级
```

```python
q_cut
```

```python
q_cut.value_counts()  # 统计分组次数
```

```python
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]  # 自定义分组区间
p_counts = pd.cut(p_change, bins)
```

```python
p_counts
```

```python
p_dummies = pd.get_dummies(data=p_counts, prefix=None)  # 得出one-hot编码矩阵, data 必须是 array-like, Series, or DataFrame, prefix 为分组前缀名
```

# 4 合并

```python
pd.concat([data, p_dummies], axis=1)  # 按行或列合并两张表的数据
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on=['key1', 'key2'])  # 使用数据库样式连接合并 DataFrame 或命名 Series 对象，默认是 inner join
```

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])  # 左链接
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='right')  # 右链接
```

```python
pd.merge(left, right, how='outer', on=['key1', 'key2'])  # 外链接
```

# 5 交叉表和透视表

#### 1 交叉表

```python
data
```

```python
time = pd.to_datetime(data.index)  # 将字符串转换为日期
```

```python
data['week'] = time.weekday  # 新增一列为日期对应的星期
```

```python
data
```

```python
data['p_n'] = np.where(data['p_change'] > 0, 1, 0)  # 将 p_change 分类，大于 0 的设置为 1，否则设置为 0
```

```python
data
```

```python
count = pd.crosstab(data['week'], data['p_n'])  # pd.crosstab 交叉表用于计算一列数据对于另外一列数据的分组个数(用于统计分组频率的特殊透视表)
```

```python
count
```

```python
count_sum = count.sum(axis=1)  # 按行计算总数
```

```python
count_sum
```

```python
per = count.div(count_sum, axis=0)  # 进行相除操作，得出比例
```

```python
per
```

```python
per.plot(kind='bar', stacked=True)  # 画出柱状图。stacked 参数为是否堆叠
```

#### 2 透视表

```python
data.pivot_table(['p_n'], index='week')  # 透视表，第一个参数为值，第二个参数为索引
```

# 6 分组和聚合

```python
col =pd.DataFrame({'color': ['white','red','green','red','green'], 'object': ['pen','pencil','pencil','ashtray','pen'],'price1':[5.56,4.20,1.30,0.56,2.75],'price2':[4.75,4.12,1.60,0.75,3.15]})
```

```python
col
```

```python
col.groupby(['color'])['price1'].mean()  # Dataframe 分组求平均值，需要聚合才可以显示（即在groupby后面要有统计方法）
```

```python
col['price1'].groupby(col['color']).mean()  # Series 分组求平均值
```

```python
col.groupby(['color'], as_index=False)['price1'].mean()  # 对于聚合输出，返回带有组标签的对象作为索引
```

### 案例：星巴克零售数据聚合

```python
starbucks = pd.read_csv("starbucks/directory.csv")
```

```python
starbucks
```

```python
country = starbucks.groupby(['Country']).count()  # 按照国家分组，求出每个国家的星巴克零售店数量
```

```python
country
```

```python
country['Brand'].sort_values(ascending=False)[:20].plot(kind='bar', figsize=(20,8))
```
