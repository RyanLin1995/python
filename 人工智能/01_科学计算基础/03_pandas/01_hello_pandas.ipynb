```python
import pandas as pd
import numpy as np
```

# 1 Pandas 使用

```python
stock_change = np.random.normal(0, 1, (10, 5))
```

```python
stock_change
```

```python
stock_rise = pd.DataFrame(stock_change)
```

```python
stock_rise
```

```python
stock_rise.shape  # 获取 Dataframe 的行和列数
```

```python
stock_code = [f'股票{i+1}' for i in range(stock_rise.shape[0])]
```

```python
stock_code
```

```python
pd.DataFrame(stock_change, index=stock_code)
```

```python
date_col = pd.date_range(start='20200403', periods=stock_rise.shape[1], freq='B')  # 返回固定频率的 DatetimeIndex
```

```python
date_col
```

```python
stock_c = pd.DataFrame(stock_change, index=stock_code, columns=date_col)
```

# 3 索引设置

### 1 修改索引值

```python
new_stock_code = [f'股票__{i+1}' for i in range(stock_rise.shape[0])]
```

```python
stock_c.index = new_stock_code  # 重新设置索引
```

```python
stock_c
```

### 2 重设索引

```python
stock_c.reset_index()  # 重设索引
```

```python
stock_c.reset_index(drop=True)  # 不保留原索引值的重设索引
```

### 3 将某列作为新索引

```python
df = pd.DataFrame({'month': [1, 4, 7, 10],
                    'year': [2012, 2014, 2013, 2014],
                    'sale':[55, 40, 84, 31]})
```

```python
df
```

```python
df.set_index('month')  # 将月设置为新索引
```

```python
df = df.set_index(['year', 'month'])  # 将月和年设置为新索引，返回的是一个具有MultiIndex的DataFrame
```

```python
df
```
